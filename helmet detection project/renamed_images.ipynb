```python

import os

def rename_images(directory):
    count = 1
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            src = os.path.join(directory, filename)
            dst = os.path.join(directory, f"Image_{count}.jpg")
            try:
                os.rename(src, dst)
                print(f'Renamed {src} to {dst}')
                count += 1
            except Exception as e:
                print(f'Error renaming {src}: {e}')

# Replace 'your_directory_path' with the path to your images directory
rename_images('/content/drive/MyDrive/analysis images')
```

```python
import os
import hashlib

# Define the folder path where you want to delete duplicate files
FOLDER_PATH = "/content/drive/MyDrive/analysis images"

def get_md5_checksum(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

def delete_file(file_path):
    try:
        os.remove(file_path)
        print(f"Deleted file: {file_path}")
    except Exception as e:
        print(f"Failed to delete file: {file_path}. Error: {e}")

def main():
    files_by_md5 = {}
    for root, dirs, files in os.walk(FOLDER_PATH):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            md5_checksum = get_md5_checksum(file_path)

            if md5_checksum in files_by_md5:
                delete_file(file_path)
            else:
                files_by_md5[md5_checksum] = file_path

if __name__ == '__main__':
    main()
```
